```python
# 행을 기준으로 처리해보기
# if 조건문 .... 
# > < >= <= ==같다.

condition = (sample_1['성별'] == '남성') #남성이 있는 행을 검색해주세요/.


# 해당 인덱스가 true 인 것만 조회를 합니다.

sample_1[condition]
```

```python


# 통합하기


# sample_1의 데이터와 code_master가 가지고 있는 데이터

# 매핑을 하면서 통합을 할 것임

# 성격이 같은 것 끼리 통합해야한다


# mapping하면서 통합

#통합한 변수 새로만들 것임



# left : 왼쪽 공간에는 sample1의 데이터를 저장해놓겠다.
# right : 오른쪽 공간에는 코드_마스터 의 데이터를저장해놓아요
# how : left ?? 기준이 되는 데이터 정의. 
#     left : 왼쪽 데이터는 전체를 조회하면서 오른쪽 데이터와 같은게 있으면 값이 조회되고. 없으면 nan 입력됨

#     right : 반대로, 오른쪽 데이터는 전체 사용 하면서, 왼쪽 데이터와 같은 게 있으면 값이 조회되고 없으면 값이 조회되지않는다.
# left_on / right_on : 왼쪽에 지정된 국적코드 값과 오른쪽의 국적코드 값이 같다. 라는 조건.
# 같은 것에 대해서만 실제 국적명이 들어가고, 같지않은 건 안들어간다. 국적코드는 조건으로 사용되는 것이고
# merge 합치는 작업을 하는데 조건으로 사용되는 변수는 국적코드이다.

# 국적코드 제외한 나머지 코드 _ 마스터 데이터는 모두 합쳐진다


sample_1_code = pd.merge(left = sample_1,   
                         right = code_master,   # 
                        how    = 'left', 
                        left_on = '국적코드',
                        right_on = '국적코드')

#left outer 조인 과 같다. 있으면 있는데로 전체, 없으면 없는대로  null 값이 들어가는 것


```

```python
sample_1_code
```

```python
# 같은 값에 대해서만- 국적코드가 연관된 값이 있는 경우에만 ... 조회하기..

#연관된 데이터가 없으면 조회하지 않기

sample_1_code = pd.merge(left = sample_1,   
                         right = code_master,   
                        how    = 'inner', 
                        left_on = '국적코드',
                        right_on = '국적코드')


# 값이 없는건 빠지고 있는 것만 출력된다 
# how inner 조건이 같은 행만 조회를 한다.
# inner join이라고 보면 된다.


```

```python
sample_1_code
```

```python
# 국적명 통합하기 - 추가할때 merge - 통합

sample_2_code = pd.merge(left = sample_2,
                        right = code_master, #왼쪽에있는 샘플2와 오른쪽에 있는 코드마스터
                        how = 'left',
                        left_on = '국적코드',
                        right_on = '국적코드') 
#판다스 안에 머지라는 함수가 있다.
```

```python
sample_2_code
```

```python
# append이용해서 행 통합하기  옆으로 통합하기보다 더 간단하다.
#python에서는 인덱스 번호가 매우 중요하다. 각각의 데이터의 index 데이터가 찍혀있따.

sample = sample_1_code.append(sample_2_code)   # 왼쪽 기준이 되는 sample, 에 sample2code를 추가하겠다 
```

```python
sample
```

```python
# index 번호 순차적으로 나열할때 ignore__indxe = True ,,, 
#  default = false이므로 . 0~5, 0~5 이렇게 나옴


sample = sample_1_code.append(sample_2_code, ignore_index=True)   # 왼쪽 기준이 되는 sample, 에 sample2code를 추가하겠다 
```

```python
sample
```
